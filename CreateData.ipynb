```
def ease_of_movement(df, n):
    """Calculate Ease of Movement for given data.
    
    :param df: pandas.DataFrame
    :param n: 
    :return: pandas.DataFrame
    """
    EoM = (df['High'].diff(1) + df['Low'].diff(1)) * (df['High'] - df['Low']) / (2 * df['Volume'])
    Eom_ma = pd.Series(EoM.rolling(n, min_periods=n).mean(), name='EoM_' + str(n))
    df = df.join(Eom_ma)
    return df
```

```
#Apple2 = yf.Ticker("AAPL")
#df1 = Apple2.history(start="2020-01-01", end='2021-01-13')
df1 = pdr.get_data_yahoo('HQY')
# shift to ensure model does not predict using current values 
#Indicators 

df1['MA'] = talib.SMA(df1['Close'],20)
df1['MA'] = df1['MA'].shift(10);
df1['EMA'] = talib.EMA(df1['Close'], timeperiod=20)
df1['EMA'] = df1['EMA'].shift(10);
df1['up_band'], df1['mid_band'], df1['low_band'] = talib.BBANDS(df1['Close'], timeperiod=20)
df1['up_band'] = df1['up_band'].shift(10);
df1['mid_band'] = df1['mid_band'].shift(10);
df1['low_band'] = df1['low_band'].shift(10);
#RSI determines whether a stock is overbought or oversold
df1['RSI'] = talib.RSI(df1['Close'], timeperiod=20)
df1['RSI'] = df1['RSI'].shift(10);
df1['MACD'], df1['macdsignal'], df1['macdhist'] = talib.MACD(df1['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
df1['MACD'] = df1['MACD'].shift(10);
df1['macdsignal'] = df1['macdsignal'].shift(10);
df1['macdhist'] = df1['macdhist'].shift(10);
df1['ATR'] = talib.ATR(df1['High'], df1['Low'], df1['Close'], timeperiod=20)
df1['ATR'] = df1['ATR'].shift(1);
df1['AD'] = talib.AD(df1['High'], df1['Low'], df1['Close'], df1['Volume'])
df1['AD'] = df1['AD'].shift(10);
df1['ADX'] = talib.ADX(df1['High'], df1['Low'], df1['Close'], timeperiod=20)
df1['ADX'] = df1['ADX'].shift(10);
df1['Previous'] = df1['Close'].shift(10);
df1['diff'] = df1['Close'].diff()
#df1 = ease_of_movement(df1,20)
up = (df1["diff"] > 0)
df1["up"] = up

df1.head()
df1 = df1.dropna()
df1.iloc[:10000]
```
